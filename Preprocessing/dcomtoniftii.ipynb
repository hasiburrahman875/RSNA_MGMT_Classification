```python
import os
import dicom2nifti

def convert_dcom_to_nifti(dicom_dir, output_dir):
    """
    Convert DICOM files to NIfTI format.

    Args:
        dicom_dir (str): Directory containing DICOM files.
        output_dir (str): Directory to save the converted NIfTI file.
    """
    # Create the output directory if it does not exist
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    # Convert DICOM to NIfTI
    dicom2nifti.convert_directory(dicom_dir, output_dir, compression=True)

# Example usage
list_dir = os.listdir("/home/mrpk9/Academic/Spring 2024/Data Mining/BraTS Dataset/BraTS Classification/train")
for i in list_dir:
    dicom_directory = f"/home/mrpk9/Academic/Spring 2024/Data Mining/BraTS Dataset/BraTS Classification/train/{i}/FLAIR"
    output_directory = f"/home/mrpk9/Academic/Spring 2024/Data Mining/BraTS Dataset/BraTS Classification/FLAIR_Niftii/{i}"
    convert_dcom_to_nifti(dicom_directory, output_directory) "take only 64 samples from dcom in regular interval to make niftii
```

```python
import os
import shutil

# Define the directory where the folders are located
directory = "/home/mrpk9/Academic/Spring 2024/Data Mining/BraTS Dataset/BraTS Classification/FLAIR_Niftii"

# Define the directory where the new files will be moved
output_directory = "/home/mrpk9/Academic/Spring 2024/Data Mining/BraTS Dataset/BraTS Classification/niftii_fewch"

# Iterate over each folder in the directory
for folder_name in os.listdir(directory):
    folder_path = os.path.join(directory, folder_name)
    
    # Check if the item in the directory is a folder
    if os.path.isdir(folder_path):
        # List all the files in the folder
        files = os.listdir(folder_path)
        
        # Iterate over each file in the folder
        for file_name in files:
            # Check if the file is a .nii.gz file
            if file_name.endswith(".nii.gz"):
                # Define the old and new file paths
                old_file_path = os.path.join(folder_path, file_name)
                #folder_name_truncate = int(folder_name)
                new_file_name =  folder_name + ".nii.gz"
                new_file_path = os.path.join(output_directory, new_file_name)
                
                # Rename the file
                os.rename(old_file_path, new_file_path)
                print(f"Renamed: {old_file_path} -> {new_file_path}")
                
        # Delete the old folder
        shutil.rmtree(folder_path)

print("Process completed successfully.")
```

```python
import os
import nibabel as nib
import numpy as np
from scipy.ndimage import zoom

def resize_nifti(input_path, output_path, new_shape, order=3):
    # Load NIfTI image
    nifti_img = nib.load(input_path)
    data = nifti_img.get_fdata()

    # Get current shape
    current_shape = data.shape

    # Calculate the zoom factor for resizing
    zoom_factor = np.divide(new_shape, current_shape)

    # Resize the image
    resized_data = zoom(data, zoom_factor, order=order)

    # Create a new NIfTI image with the resized data
    new_nifti_img = nib.Nifti1Image(resized_data, affine=nifti_img.affine, header=nifti_img.header)

    # Save the resized NIfTI image
    nib.save(new_nifti_img, output_path)
    print("Resized image saved to:", output_path)

def resize_nifti_folder(input_folder, output_folder, new_shape, order=3):
    # Create the output folder if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # List all NIfTI files in the input folder
    nifti_files = [f for f in os.listdir(input_folder) if f.endswith('.nii') or f.endswith('.nii.gz')]

    # Resize each NIfTI image
    for nifti_file in nifti_files:
        input_path = os.path.join(input_folder, nifti_file)
        output_path = os.path.join(output_folder, nifti_file)
        resize_nifti(input_path, output_path, new_shape, order=order)

# Example usage
input_folder = "/home/mrpk9/Academic/Spring 2024/Data Mining/BraTS Dataset/BraTS Classification/only_niftii"
output_folder = "/home/mrpk9/Academic/Spring 2024/Data Mining/BraTS Dataset/BraTS Classification/test_channel"
new_shape = (256, 256, 64)  # Specify the new shape here
resize_nifti_folder(input_folder, output_folder, new_shape)
```
